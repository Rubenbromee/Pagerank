```python
import numpy as np
from numpy.linalg import eig
import scipy.sparse.linalg

data = open('pageRank-gr0.epa.txt', 'r')

a = True
edge_list = np.empty((8965, 2), dtype=int)
i = 0
M=np.zeros((4772,4772), dtype=float) 

while a:
    data_line = data.readline()
    if not data_line:
        print("eof")
        a = False
        break
    if data_line[0] == 'e':
        t = data_line.split()
        edge_list[i][0] = t[1]
        edge_list[i][1] = t[2]
        i = i + 1

print(i)
# Creating link matrix, binary
for i in range(len(edge_list)):
    M[edge_list[i][1]][edge_list[i][0]] = 1

column_sum = M.sum(axis=0)

# Create d
d = np.zeros(len(column_sum))
for x in range(len(column_sum)):
 if(np.sum(M[:, x]) == 0):
     d[x] = 1

# To prevent division with 0
for x in range(len(column_sum)):
    if column_sum[x] == 0:
        column_sum[x] = 1

for i in range(len(column_sum)):
    if np.sum(M[:, i]) != 0:
        M[:, i] = M[:, i] / np.sum(M[:, i])
    #print(np.sum(M[:, i]))
```

```python
e = np.ones(len(column_sum))
ed = np.outer(e,d)
P = M + 1/len(d) * ed

alpha = 0.85
eeT = np.outer(e,e)
A = alpha * P + (1 - alpha) * 1/len(column_sum) * eeT

eigs_A = eig(A)
print(eigs_A)

```

```python
column_sum_A = A.sum(axis=0) 
for i in column_sum_A:
    print(i)
```

```python
count=0
residual = 1
z = np.ones(M.shape[1]) / (M.shape[1])
v = np.ones(M.shape[1]) / (M.shape[1])
alpha = 0.85
res = 1
error = []
error_sc = 1
while count < 100:
    yhat = alpha * np.matmul(M,z)
    beta = 1 - np.linalg.norm(yhat, ord=1)
    y = yhat + beta * v
    res = np.linalg.norm(y - z, ord=1)
    error_sc = np.linalg.norm(y - eigs_A[1], ord=1) / len(y)
    error.append(error_sc)
    z = y
    print('Error_sc = ', error_sc)
    count=count+1
print(y)
```

```python
import matplotlib.pyplot as plt
plt.clf()
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_ylabel('Error')
ax.set_xlabel('Iteration')
plt.plot(error)
plt.show()
```
